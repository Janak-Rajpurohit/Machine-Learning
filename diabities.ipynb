```python
import pandas as pd 
import numpy as np
```

```python
df=pd.read_csv("diabetes.csv")
df.head()
```

```python
df.sample(20)
```

```python
df[['Glucose'	,'BloodPressure'	,'SkinThickness',	'Insulin'	,'BMI'	,'DiabetesPedigreeFunction',	'Age']]=df[['Glucose'	,'BloodPressure'	,'SkinThickness',	'Insulin'	,'BMI'	,'DiabetesPedigreeFunction',	'Age']].replace(0,np.NaN)
```

```python
df.isnull().sum()
```

```python
df['Glucose']=df['Glucose'].replace(np.NaN,df["Glucose"].mean())
df["BloodPressure"]=df['BloodPressure'].replace(np.NaN,df["BloodPressure"].mean())
df['SkinThickness']=df["SkinThickness"].replace(np.NaN,df["SkinThickness"].mean())
df["Insulin"]=df["Insulin"].replace(np.NaN,df["Insulin"].mean())
df["BMI"]=df["BMI"].replace(np.NaN,df["BMI"].mean())
df["DiabetesPedigreeFunction"]=df["DiabetesPedigreeFunction"].replace(np.NaN,df["DiabetesPedigreeFunction"].mean())
df["age"]=df["Age"].replace(np.NaN,df["Age"].mean())
df.isnull().sum()
```

```python
## models
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
## for accuracy
from sklearn.model_selection import cross_val_score 
```

```python
y=df["Outcome"]
y
```

```python
x=df.drop(["Outcome"],axis=1)
x
```

```python
from sklearn.model_selection import train_test_split
x_train ,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)

from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
x_train
```

```python
def get_score(model,xtrain,ytrain,xtest,ytest):
    model.fit(xtrain,ytrain)
    return model.score(xtest,ytest)
```

```python
print(get_score(LogisticRegression(),x_train,y_train,x_test,y_test))
print(get_score(SVC(),x_train,y_train,x_test,y_test))
print(get_score(RandomForestClassifier(),x_train,y_train,x_test,y_test))
print(get_score(DecisionTreeClassifier(max_depth=5,random_state=40),x_train,y_train,x_test,y_test))
```

```python
print(cross_val_score(LogisticRegression(),x,y))
print(cross_val_score(SVC(),x,y))
print(cross_val_score(RandomForestClassifier(),x,y))
```
