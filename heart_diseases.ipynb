```python
import pandas as  pd 
```

```python
df=pd.read_csv("heart.csv")
df
```

```python
col=df.select_dtypes(include="object").columns
df=pd.get_dummies(df,columns=col)
df
```

```python
x=df.drop(["HeartDisease"],axis=1)
x
```

```python
y=df["HeartDisease"]
y
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier

from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25)

len(x_train) , len(x_test)
```

```python
def get_score(model,xtrain,ytrain,xtest,ytest):
    model.fit(xtrain,ytrain)
    return model , model.score(xtest,ytest)
```

```python
print(get_score(LogisticRegression(),x_train,y_train,x_test,y_test))
print(get_score(RandomForestClassifier(),x_train,y_train,x_test,y_test))
print(get_score(SVC(),x_train,y_train,x_test,y_test))
print(get_score(DecisionTreeClassifier(),x_train,y_train,x_test,y_test))
```

```python
from sklearn.model_selection import cross_val_score

print(sum(cross_val_score(LogisticRegression(),x,y))/5)
print(sum(cross_val_score(RandomForestClassifier(),x,y))/5)
print(sum(cross_val_score(SVC(),x,y))/5)
print(sum(cross_val_score(DecisionTreeClassifier(),x,y))/5)
```

Random forest & Logistic reg is performing well  

```python
from sklearn.model_selection import GridSearchCV

lr=GridSearchCV(LogisticRegression(),{
    'solver':['newton-cg','lbfgs','liblinear','sag','saga']
},cv=5,return_train_score=False)

lr.fit(x,y)
lr.best_score_ ,lr.best_params_
```

```python
rf=GridSearchCV(RandomForestClassifier(),{
    'n_estimators':[1,10,20,30,40,50],
    'max_depth':[20,30,35,40,45,50]
},cv=5,return_train_score=False)

rf.fit(x,y)
rf.best_score_,rf.best_params_
```

**random forest and logistic are performing at same level with score and param of**  

*logistic reg (0.8332858161083394, {'solver': 'lbfgs'})* <br />
*random forest (0.8376098835827988, {'max_depth': 30, 'n_estimators': 40})*
