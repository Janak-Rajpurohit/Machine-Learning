```python
import pandas as pd 

from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV

from sklearn.linear_model import LogisticRegression
from sklearn.svm  import SVC
from sklearn.svm import LinearSVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier

```

```python
df=pd.read_csv("diabetes_012_health_indicators_BRFSS2015.csv")
df.head()
```

```python
x=df.drop(["Diabetes_012","Education","Income"],axis=1)
x
```

```python
y=df[["Diabetes_012"]]
y
```

```python
def get_score(model,xtrain,ytrain,xtest,ytest):
    model.fit(xtrain,ytrain)
    return model , model.score(xtest,ytest)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20)
len(x_train),len(x_test)
```

```python
print(get_score(LogisticRegression(),x_train,y_train,x_test,y_test))
# print(get_score(SVC(probability=False),x_train,y_train,x_test,y_test))  it takes tooo long to train data >10000
# print(get_score(LinearSVC(random_state=0, tol=1e-5),x_train,y_train,x_test,y_test))
print(get_score(RandomForestClassifier(),x_train,y_train,x_test,y_test))
print(get_score(DecisionTreeClassifier(),x_train,y_train,x_test,y_test))
```

```python
print(sum(cross_val_score(LogisticRegression(),x,y))/5)
print(sum(cross_val_score(RandomForestClassifier(),x,y))/5)
print(sum(cross_val_score(DecisionTreeClassifier(),x,y))/5)
```

```python
clf=GridSearchCV(LogisticRegression(),{
    "solver":["liblinear","lbfgs","saga"],
    "C":[10,20,30]
})
clf.fit(x,y)
clf.best_score_ , clf.best_params_
```

```python
clf=GridSearchCV(RandomForestClassifier(),{
    "n_estimators":[1,10,15,20,30,40]
})
clf.fit(x,y)
clf.best_score_ , clf.best_params_
```
